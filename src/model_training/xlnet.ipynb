```python
# train_file_path = '/content/gdrive/My Drive/Colab Notebooks/raw_data/fulltrain.csv'
train_file_path = '/content/gdrive/My Drive/Colab Notebooks/cs4248/Project/fulltrain.csv'
#test_file_path = '/content/gdrive/My Drive/Colab Notebooks/raw_data/balancedtest.csv'
test_file_path = '/content/gdrive/My Drive/Colab Notebooks/cs4248/Project/balancedtest.csv'
model_path = '/content/gdrive/My Drive/Colab Notebooks/cs4248/Project/outputs' #change this to the outputs folder path
  model = ClassificationModel('xlnet', model_path, num_labels=4, args={"overwrite_output_dir": True, "num_train_epochs":4, "do_lower_case":True})

train_df = pd.read_csv(train_file_path, names=['label', 'text'])
test_df = pd.read_csv(test_file_path, names=['label', 'text'])

train_df = train_df.astype({'label':'int'}) #model label range is from 0-3 instead of 1-4 so some preprocessing is needed
label = train_df['label'].copy()
label = label-1
text = train_df['text'].copy()
new = pd.DataFrame({'text':text,'label':label})
train_df = new.copy()

train_df.head()
```

```python
test_df = test_df.astype({'label':'int'}) #preprocessing for test dataset
label = test_df['label'].copy()
label = label - 1
text = test_df['text'].copy()
new = pd.DataFrame({'text':text,'label':label})
test_df = new.copy()

test_df.head()
```

Prediction

```python
X_test = list(test_df['text'])
Y_test = test_df['label']
y_pred = model.predict(X_test)
score = f1_score(Y_test,y_pred[0], average = 'macro')
print('score on test set = {}'.format(score))
```
